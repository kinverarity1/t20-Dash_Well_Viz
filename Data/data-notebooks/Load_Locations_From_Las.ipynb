```python
def files_to_read(list_of_files):
    data = list()

    for f in list(list_of_files):

        las = lasio.read(f)
        well = las.well['WELL']['value']
        lon = las.well['LONG']['value']
        lat = las.well['LATI']['value']
        
        d = dict(well=well, lon=lon, lat=lat)
        data.append(d)
    
    return pd.DataFrame(data)
        
```

```python
df = files_to_read(Path('../').glob('*Decim.LAS'))
df
```

```python
lon_dms = df['lon'].str.split('/', expand=True).astype('float').rename({0:'Lon_D', 1:'Lon_M', 2:'Lon_S'}, axis=1)
lat_dms = df['lat'].str.split('/', expand=True).astype('float').rename({0:'Lat_D', 1:'Lat_M', 2:'Lat_S'}, axis=1)

joined_df = pd.concat([df, lat_dms, lon_dms,], axis=1)
```

```python
las_header = joined_df.sort_values('well').reset_index(drop='True')
las_header
```
